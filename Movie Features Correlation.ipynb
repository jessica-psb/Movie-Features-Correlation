```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
#Read data
df = pd.read_csv('Downloads/MoviesData/Movies.csv')
```

```python
#Look at the data
df.head()
```

```python
# Check missing data percentage for each column
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Drop rows with missing values
df = df.dropna(how='any',axis=0)
```

```python
# Data types of the columns
df.dtypes
```

```python
# Change data type of columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
# Correcting false 'year' from 'released' year
df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
# Order rows by gross
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python
# Check any duplicates
df.drop_duplicates()
```

```python
# See budget correlation with gross 
# Scatter plot with budget vs gross
matplotlib.rcParams['figure.figsize'] = (12, 8) #Adjust the config of the plots we will create
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show
```

```python
# Determine if it's correlated
#Plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"},  line_kws={"color":"blue"})
```

```python
# Looking at correlation
df.corr(method='pearson') #pearson, kendall, spearman
```

```python
# High correlation between budget and gross
```

```python
# Visualize the correlation using heatmap
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movies Features')
plt.show()
```

```python
# See company correlation with gross 
# Numerize columns with object dtype
df_numerized = df.copy()
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
# Compare the numerized dataframe with the original dataframe to check if it's right
df.head()
```

```python
# Check the company correlation with gross earnings using heatmap
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movies Features')
plt.show()
```

```python
# Unstack correlation matrix into pairs
corr_matrix = df_numerized.corr()
corr_pairs = corr_matrix.unstack()
corr_pairs
```

```python
# Order correlation pairs
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# High correlation pairs
high_corr = sorted_pairs[sorted_pairs > 0.5]
high_corr
```
